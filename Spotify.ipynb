```python
import time
def performance(fn):
    def wrapper(*args, **kwargs):
        t1 = time.time()
        result = fn(*args, **kwargs)
        t2 = time.time()
        print(f'took {t2-t1} s')
        return result
    return wrapper
```

```python
@performance
def thirdway(df):
    df.drop(df.index[df['stream30s'] == 0], inplace=True)
thirdway(df2)
```

```python
idea = finaldf.copy()
idea = idea[(idea['stream30s'] > idea['stream30s'].mean()) & \
            (idea['mau_both_months'] > idea['mau_both_months'].mean()) & \
            (idea['skippers'] > idea['skippers'].mean())]
idea['True Listens'] = idea['stream30s'] - idea['monthly_owner_stream30s']
idea.drop(['monthly_owner_stream30s','stream30s'],axis=1,inplace=True)
idea.sort_values('mau_both_months',inplace=True,ascending=False)
idea.reset_index(inplace=True,drop=True)
idea.head(60)
```

```python
idea['n_artists'].mean(), idea['n_tracks'].mean(), idea['skippers'].mean()
```

```python
mood = idea.groupby('mood_1')['True Listens'].mean()
mood = pd.DataFrame(mood)
mood.sort_values('True Listens', inplace=True, ascending=False)
mood.reset_index(inplace=True)
mood
```

```python
fig,ax=plt.subplots(1,1,figsize=(35,25))
sns.barplot(data=mood,x='mood_1',y='True Listens')
```

```python
genre = idea.groupby('genre_1')['True Listens'].mean()
genre = pd.DataFrame(genre)
genre.sort_values('True Listens', inplace=True, ascending=False)
genre.reset_index(inplace=True)
genre
```

```python
fig,ax=plt.subplots(1,1,figsize=(35,25))
sns.barplot(data=genre,x='genre_1',y='True Listens')
```
